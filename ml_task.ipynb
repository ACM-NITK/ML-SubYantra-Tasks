Now, in several instances columns containing numeric data can have “object” data type. Some special characters like ? or space can be present amidst continuous variables because of lack of data points. We will check the unique entries in such cases to remove them and change the data types:

```python

for column in columns :
    uni_val_col = df[column].unique()
    print ('The unique values in ' , column , 'are ', uni_val_col)
```

Creating separate continuous and categorical dataframes

```python
## Create a dataframe with continuous columns 
df_cont=df.select_dtypes(include=['int64','float64'])
df_int = df.select_dtypes(include = ['int64']) 
df_float = df.select_dtypes(include = ['float64'])
## Create a dataframe with categorical columns 
df_cat = df.select_dtypes(include =['object'])
```

```python
print(df_int.head())
print(df_float.head())
print('categorical dataframe')
print(df_cat.head())
```

We should check for similar special characters  in categorical columns and replace them with appropriate values.


```python
df.replace(to_replace=["?", ";"], value=np.nan)
```

Get the number of missing data points per column and look at the numbvr of missing points in the first ten columns.

```python
missing_values_count = df.isnull().sum()
missing_values_count[0:9]
```

It might be helpful to see what percentage of the values in our dataset were missing 

```python
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100
print('percent missing=',percent_missing)
```

We can use the Panda's fillna() function to fill in missing values in a dataframe for us.

```python
df_cont.fillna(537.870553	)
```

Percentage of outliers and other descriptive statistical measures

```python

df_outliers = pd.DataFrame(index=df_cont.columns, columns=['outliers', 'outliers%']) 
for col in df_cont.columns:
    if any(x in str(df_cont[col].dtype) for x in ['int', 'float']):
        df_outliers.loc[col, 'count'] = len(df_cont)
        df_outliers.loc[col, 'q1'] = df_cont[col].quantile(0.25)
        df_outliers.loc[col, 'q3'] = df_cont[col].quantile(0.75)
        df_outliers.loc[col, 'iqr'] = df_outliers.loc[col, 'q3'] - df_outliers.loc[col, 'q1']
        df_outliers.loc[col, 'lower'] = df_outliers.loc[col, 'q1'] - (3 * df_outliers.loc[col, 'iqr'])
        df_outliers.loc[col, 'upper'] = df_outliers.loc[col, 'q3'] + (3 * df_outliers.loc[col, 'iqr'])
        df_outliers.loc[col, 'min'] = df[col].min()
        df_outliers.loc[col, 'max'] = df[col].max()
        df_outliers.loc[col, 'outliers'] = ((df_cont[col] < df_outliers.loc[col, 'lower']) | (df[col] > df_outliers.loc[col,'upper'])).sum()
        df_outliers.loc[col, 'outliers%'] = np.round(df_outliers.loc[col,'outliers'] / len(df_cont) *100)
        print(df_outliers)
```

Treating outliers

```python
from scipy import stats as st
z_scores = st.zscore(df_cont)
abs_z_scores = np.abs(z_scores)
entries = (abs_z_scores < 3).all(axis=1)
df_cont = df_cont[entries]
```

Saving preprocessed dataframe in a csv  

```python
df_cont.to_csv('preprocessed_data.csv')
```
