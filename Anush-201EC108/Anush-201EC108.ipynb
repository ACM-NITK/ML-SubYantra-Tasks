```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_housingprices.csv')
```

```python
df.info()
```

```python
# null values
df.isna().sum()
```

```python
## null values in 'total_bedrooms' percentage of null values
df.isnull().mean().mean()*100
```

```python
## null value percentage is very low we can drop null values
df.dropna(inplace=True)
df.info()
```

```python
## dropping duplicates 
df.drop_duplicates(inplace=True)
df.info()
```

```python
## visualising data
df.describe()
```

```python
df.corr()
```

```python
#Data distribution plots
for x in df.columns:
 sns.displot(df[x],kde=True)
```

```python
#correlation scatter plots
for x in df.columns:
 for y in df.columns:
    if(x==y):
        break
    df.plot.scatter(x,y)
```

```python
#correlation heat map
sns.heatmap(df.corr(), annot=True)
```
